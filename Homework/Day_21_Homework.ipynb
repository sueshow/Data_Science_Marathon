````python
# 載入數據集
tips = sns.load_dataset('tips')
fmri = sns.load_dataset('fmri')

print('tips = ', tips.head(5), '\n \n')
print('fmri = ', fmri.head(5), '\n \n')
````

````python
#設定圖形樣式
sns.set_style('whitegrid')
sns.relplot(x='total_bill', y='tip', hue='sex', style='time', data=tips)
````

【進階】

* 作業(1)：繪製模型殘差圖進一步了解不同的特徵值所呈現的關係圖

````python
# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
np.random.seed(42)
x = np.random.normal( 2, 1, 75 )
y = 2 + 1.5 * x + np.random.normal( 0, 2, 75 )

print( 'x = \n', x, '\n' )
print( 'y = \n', y )
````

````python
# 畫圖
plt.figure(figsize=(8, 5))
sns.residplot( x=x, y=y, lowess=True, color='g' )
plt.show()
````

````python
sns.distplot(x, kde=False, bins=20 )
plt.show()
````

````python
sns.histplot( x, bins=20, kde=False )
plt.show()
````

````python
sns.histplot( x, bins=20, kde=True )
plt.show()
````

````python
sns.histplot( x, bins=40, kde=True )
plt.show()
````
