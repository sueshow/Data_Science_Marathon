```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

【基礎題】

```python
# 設定要畫的 x, y 數據 list....
x = np.arange(0,180)
y = np.cos( x*np.pi/180.0 )
print( 'x= \n', x, '\n' )
print( 'y= \n', y )
```

```python
# 畫圖
plt.plot(x,y)
plt.xlim(0, 200)
plt.ylim(-1, 1.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('sin')
plt.show() 
```

#### Sin

```python
# 畫出完整的 sin 圖形
x = np.arange( 0, 2*np.pi, 0.1 )
y_sin = np.sin(x)

plt.plot( x, y_sin )
plt.show()
```

#### Cos

```python
#1. 畫出完整的 cos 圖形

x = np.arange( 0, 2*np.pi, 0.1 )
y_cos = np.cos(x)

plt.plot( x, y_cos )
#plt.show()
plt.savefig('D:/資料分析彙整/Project/Data Science Marathon/data/Homework_018.png',dpi=300,format='png')
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

print( 'X= \n', X, '\n' )
print( 'Y= \n', Y )
```

#### Scatter Plots

```python
#2. 給定散點圖顏色
plt.scatter( X, Y, s=75, c=np.sin(X/Y), alpha=.4)
#plt.scatter(X, Y, c='red', marker='o', alpha=0.4, linewidth=1.0, edgecolor='black')

plt.title('Scatter plot')
#plt.legend( loc='upper right')
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```

【進階題】

```python
#1. 畫出箱型圖，並判斷哪個欄位的中位數在 300~400 之間？

data = pd.read_csv('D:/資料分析彙整/Project/Data Science Marathon/data/boston.csv')
data = pd.DataFrame(data)
data
```

```python
data.boxplot(figsize = (15, 6))
plt.xticks(rotation = 'vertical')
plt.ylim(0, 800)
```

```python
median_var = data.median().between(300, 400)
print(median_var)
```

```python
print('AX 及 B 欄位的中位數位於300~400之間')
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係？

x = data['NOX']
y = data['DIS']
plt.scatter(x, y, color='b', alpha=0.3, marker='o')
plt.xlabel('NOX')
plt.ylabel('DIS')
plt.title('NOX vs DIS')
plt.show()
```
